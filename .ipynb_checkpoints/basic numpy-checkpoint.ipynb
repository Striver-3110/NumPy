# Slicing in 1 dimentional array

```python
import numpy as np
a=[]
size = int(input("enter size"))
for i in range(size):
    a.append(int(input("enter number")))

numpyarray = np.array(a)
myarray = numpyarray[:size+1]
print(myarray)
myarray = numpyarray[0:]
print(myarray)
myarray = numpyarray[1:3]
print(myarray)
```

# 2D array in numpy

```python
# creating 2d array

import numpy as np
r = int(input("enter how many rows"))
c = int(input("enter how many columns"))
a = []
for i in range(r):
    row = []
    for j in range(c):
        num = int(input("enter element"))
        row.append(num)
    a.append(row)
a = np.array(a)
a.reshape(r,c)
for i in range(r):
    for j in range(c):
        print(a[i][j])#double bracket notation
for i in range(r):
    for j in range(c):
        print(a[i,j])#single bracket notation
        
print(a)
```

slicing in 2d array

# view and copy of an array

```python
import numpy as np
a=np.array([1,2,3,4,5,6,7])
slic = a[0:5].view()
# slic = a[0:5] # this line is same as above line 

slic[:]=0
print(a)
```

```python
import numpy as np
a=np.array([1,2,3,4,5,6,7])
slic = a[0:5].copy()
slic[:]=0
print(a)
```
